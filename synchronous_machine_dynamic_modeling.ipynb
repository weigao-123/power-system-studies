From Kirchhoff's Voltage Law, Ohm's Law, Faraday's Law, Newton's Second Law time domain, unscaled, we have:

Stator:
$$
v_a = i_a r_s + \frac{d\lambda_a}{dt}
$$

$$
v_b = i_b r_s + \frac{d\lambda_b}{dt}
$$

$$
v_c = i_c r_s + \frac{d\lambda_c}{dt}
$$

Rotor:
$$
v_{fd} = i_{fd} r_{fd} + \frac{d\lambda_{fd}}{dt}
$$

$$
v_{1d} = i_{1d} r_{1d} + \frac{d\lambda_{1d}}{dt}
$$

$$
v_{1q} = i_{1q} r_{1q} + \frac{d\lambda_{1q}}{dt}
$$

$$
v_{2q} = i_{2q} r_{2q} + \frac{d\lambda_{2q}}{dt}
$$

Shaft:
$$
\frac{d\theta_{\text{shaft}}}{dt} = \frac{2}{P} \omega
$$

$$
J \frac{2}{P} \frac{d\omega}{dt} = T_m - T_e - T_{f\omega}
$$

The stator abc variables are sinusoidal in steady state.

This oscillatory equilibrium can be made constant through a transformation.

Transform the state variables $v$, $i$, and $\lambda$ onto the rotor.

Transformation of Stator abc Variables to dq0

$$
\begin{bmatrix}
v_d \\
v_q \\
v_o
\end{bmatrix}
\triangleq T_{dq0}
\begin{bmatrix}
v_a \\
v_b \\
v_c
\end{bmatrix}
\quad \text{or} \quad i, \lambda
$$

Inverse Transformation from dq0 to abc

$$
\begin{bmatrix}
v_a \\
v_b \\
v_c
\end{bmatrix}
=
T^{-1}_{dq0}
\begin{bmatrix}
v_d \\
v_q \\
v_o
\end{bmatrix}
$$


The transformation matrix is not unique, depends on the choice of the reference.

Forward Transformation Matrix \( T_{dq0} \)

$$
T_{dq0} \triangleq \frac{2}{3}
\begin{bmatrix}
\sin\left(\frac{P}{2}\theta_{\text{shaft}}\right) & \sin\left(\frac{P}{2}\theta_{\text{shaft}} - \frac{2\pi}{3}\right) & \sin\left(\frac{P}{2}\theta_{\text{shaft}} + \frac{2\pi}{3}\right) \\
\cos\left(\frac{P}{2}\theta_{\text{shaft}}\right) & \cos\left(\frac{P}{2}\theta_{\text{shaft}} - \frac{2\pi}{3}\right) & \cos\left(\frac{P}{2}\theta_{\text{shaft}} + \frac{2\pi}{3}\right) \\
\frac{1}{2} & \frac{1}{2} & \frac{1}{2}
\end{bmatrix}
$$

Inverse Transformation Matrix \( T^{-1}_{dq0} \)

$$
T^{-1}_{dq0} = 
\begin{bmatrix}
\sin\left(\frac{P}{2}\theta_{\text{shaft}}\right) & \cos\left(\frac{P}{2}\theta_{\text{shaft}}\right) & 1 \\
\sin\left(\frac{P}{2}\theta_{\text{shaft}} - \frac{2\pi}{3}\right) & \cos\left(\frac{P}{2}\theta_{\text{shaft}} - \frac{2\pi}{3}\right) & 1 \\
\sin\left(\frac{P}{2}\theta_{\text{shaft}} + \frac{2\pi}{3}\right) & \cos\left(\frac{P}{2}\theta_{\text{shaft}} + \frac{2\pi}{3}\right) & 1
\end{bmatrix}
$$


We have new stator variables $v_d, v_q, v_o, i_d, i_q, i_o, \lambda_d, \lambda_q, \lambda_o$, 
the rotor shaft speed $\omega$ = $d\theta_{\text{shaft}}/dt$,

by substituting $v_{abc} = T^{-1}_{dq0}v_{dq0}$, $i_{abc} = T^{-1}_{dq0}i_{dq0}$, $\lambda_{abc} = T^{-1}_{dq0}\lambda_{dq0}$ back into the stator equations in abc coordinates, we have:


Stator in dq0 coordinates:
$$
v_d = i_d r_s + \frac{d\lambda_d}{dt} - \omega \lambda_q
$$

$$
v_q = i_q r_s + \frac{d\lambda_q}{dt} + \omega \lambda_d
$$

$$
v_o = i_o r_s + \frac{d\lambda_o}{dt}
$$

### Single-Machine Steady State

with all the state derivtaves being zero for steady-state

we have $\omega = \omega_s$ from Eq. (3.155), then for Eq. (3.148) and Eq. (3.149), we have

$$
0 = R_s I_d + \phi_q + V_d \tag{3.215}
$$

$$
0 = R_s I_q - \phi_d + V_q \tag{3.216}
$$

and also, $\phi_{1d} = E'_q - (X'_d - X_{ls}) I_d$ from Eq. (3.152)

then for Eq. (3.151), we have

$$
0 = -E'_q - (X_d - X'_d) I_d + E_{fd}
$$

for Eq. (3.157), we have

$$
\phi_d = E'_q - X'_d I_d
$$

and then, $\phi_{2q} = E'_d - (X'_q - X_{ls}) I_q$ from Eq. (3.154)

then for Eq. (3.153), we have

$$
0 = -E'_d + (X_q - X'_q) I_q
$$

for Eq. (3.158), we have

$$
\phi_q = -E'_d - X'_q I_q
$$

then, substituting for $\phi_d$ and $\phi_q$ in Eq. (3.215) and Eq. (3.216) gives

$$
V_d = -R_s I_d + E'_d + X'_q I_q \tag{3.224}
$$

$$
V_q = -R_s I_q + E'_q - X'_d I_d \tag{3.225}
$$

These two real algebraic equations can be written as one complex equation

$$
(V_d + jV_q)e^{j(\delta - \pi/2)} = -(R_s + jX_q) (I_d + jI_q) e^{j(\delta - \pi/2)} +  \overline{E} \tag{3.226}
$$
where (further simplified using Eq. (3.151))
$$
\overline{E} = \left[(E'_d - (X_q - X'_q) I_q) + j(E'_q + (X_q - X'_d) I_d)\right] e^{j(\delta - \pi/2)} \\
= j\left[(X_q - X'_d) I_d + E'_q \right] e^{j(\delta - \pi/2)} \\
= j\left[(X_q - X_d) I_d + E_{fd} \right] e^{j(\delta - \pi/2)} \\
= \left[(X_q - X_d) I_d + E_{fd} \right] e^{j\delta} \tag{3.227}
$$


This give a considerable physical significance to the circuit, in which $\overline{E}$ is the "internal" voltage, and the power angle $\delta$.

Thus

$$
\overline{E} = \overline{V_t} + (R_s + jX_q) \overline{I}
$$

0 = (R_s + R_e) I_q + (X'_d + X_{ep}) I_d - E'_q + V_s \cos(\delta - \theta_{vs})

Note, usually, in real system, the machine initial condition is determined in power flow study, and usually as "PV" bus, thus, the inputs are V_t magnitude and P_M real power, V_s magnitude and angle

The modified two_axis model is as follows:

### Two-Axis Model with power flow results as inputs

$$
T'_{do} \frac{dE'_q}{dt} = -E'_q - (X_d - X'_d) I_d + E_{fd}
$$

$$
T'_{qo} \frac{dE'_d}{dt} = -E'_d + (X_q - X'_q) I_q
$$

$$
\frac{d\delta}{dt} = \omega - \omega_s
$$

$$
\frac{2H}{\omega_s} \frac{d\omega}{dt} = P_M / \omega - E'_d I_d - E'_q I_q - (X'_q - X'_d) I_d I_q - T_{FW}
$$


$$
0 = (R_s + R_e) I_d - (X'_q + X_{ep}) I_q - E'_d + V_s \sin(\delta - \theta_{vs})
$$

$$
0 = (R_s + R_e) I_q + (X'_d + X_{ep}) I_d - E'_q + V_s \cos(\delta - \theta_{vs})
$$


$$
V_d = R_e I_d - X_{ep} I_q + V_s \sin(\delta - \theta_{vs})
$$

$$
V_q = R_e I_q + X_{ep} I_d + V_s \cos(\delta - \theta_{vs})
$$

$$
V_d = V_t \sin(\delta - \theta_{vt})
$$

$$
V_q = V_t \cos(\delta - \theta_{vt})
$$

```python
import numpy as np
from scipy.optimize import fsolve

def two_axis_model_steady_state(y, machine_params, network_params, power_flow_results):
  r"""
  Two-Axis Model for synchronous machine connected with an infinity bus
  
  $$
  T'_{do} \frac{dE'_q}{dt} = -E'_q - (X_d - X'_d) I_d + E_{fd}
  $$

  $$
  T'_{qo} \frac{dE'_d}{dt} = -E'_d + (X_q - X'_q) I_q
  $$

  $$
  \frac{d\delta}{dt} = \omega - \omega_s
  $$

  $$
  \frac{2H}{\omega_s} \frac{d\omega}{dt} = P_M / (omega / omega_s) - E'_d I_d - E'_q I_q - (X'_q - X'_d) I_d I_q - T_{FW}
  $$

  $$
  0 = (R_s + R_e) I_d - (X'_q + X_{ep}) I_q - E'_d + V_s \sin(\delta - \theta_{vs})
  $$

  $$
  0 = (R_s + R_e) I_q + (X'_d + X_{ep}) I_d - E'_q + V_s \cos(\delta - \theta_{vs})
  $$

  $$
  V_d = R_{ed} I_d - X_{ep} I_q + V_s \sin(\delta - \theta_{vs})
  $$

  $$
  V_q = R_{eq} I_q + X_{ep} I_d + V_s \cos(\delta - \theta_{vs})
  $$

  $$
  V_d = V_t \sin(\delta - \theta_{vt})
  $$

  $$
  V_q = V_t \cos(\delta - \theta_{vt})
  $$
  """
  
  E_q_prime, E_d_prime, E_fd, V_d, V_q, I_d, I_q, delta, omega, theta_vt = y

  H, R_s, X_d, X_q, X_d_prime, X_q_prime, T_do_prime, T_qo_prime, omega_s, T_FW = machine_params["H"], machine_params["R_s"], machine_params["X_d"], machine_params["X_q"], machine_params["X_d_prime"], machine_params["X_q_prime"], machine_params["T_do_prime"], machine_params["T_qo_prime"], machine_params["omega_s"], machine_params.get("T_FW", 0)

  R_e, X_ep = network_params["R_e"], network_params["X_ep"]

  V_s, V_t, P_M = power_flow_results["V_s"], power_flow_results["V_t"], power_flow_results["P_M"]
  theta_vs = np.angle(V_s)
  V_s = np.abs(V_s)
  
  eq1 = -E_q_prime - (X_d - X_d_prime) * I_d + E_fd
  eq2 = -E_d_prime + (X_q - X_q_prime) * I_q
  eq3 = omega - omega_s
  eq4 = P_M / (omega / omega_s) - E_d_prime*I_d - E_q_prime*I_q - (X_q_prime - X_d_prime) * I_d * I_q - T_FW
  eq5 = (R_s + R_e) * I_d - (X_q_prime + X_ep) * I_q - E_d_prime + V_s * np.sin(delta - theta_vs)
  eq6 = (R_s + R_e) * I_q + (X_d_prime + X_ep) * I_d - E_q_prime + V_s * np.cos(delta - theta_vs)
  eq7 = -V_d + R_e*I_d - X_ep*I_q + V_s * np.sin(delta - theta_vs)
  eq8 = -V_q + R_e*I_q + X_ep*I_d + V_s * np.cos(delta - theta_vs)
  eq9 = -V_d + V_t * np.sin(delta - theta_vt)
  eq10 = -V_q + V_t * np.cos(delta - theta_vt)

  return [eq1, eq2, eq3, eq4, eq5, eq6, eq7, eq8, eq9 ,eq10]
```

```python
machine_params = {
  "H": 3.0,
  "R_s": 0,
  "X_d": 2.1,
  "X_q": 2.0,
  "X_d_prime": 0.3,
  "X_q_prime": 0.5,
  "T_do_prime": 7.0,
  "T_qo_prime": 0.75,
  "omega_s": 2*np.pi*60,
  "T_FW": 0,
}

network_params = {
  "R_e": 0,
  "X_ep": 0.20 * (0.10 + 0.20) / (0.20 + (0.10 + 0.20)) + 0.10,
}
power_flow_results = {
  "V_s": 1.0 + 1j*0,
  "V_t": 1.0946,
  "P_M": 1.0,
}

# steaty_state = [1.130, 0.533, 2.913, 0.7107, 0.8326, 0.9909, 0.3553, 52.1/180*np.pi, 2*np.pi*60, 0.2024]
initial_guess = [0] * 10
initial_guess[-2] = 2*np.pi*60  # initialize the omega with non-zero to avoid dividing by zero warnings
solution = fsolve(two_axis_model_steady_state, initial_guess, args=(machine_params, network_params, power_flow_results))
E_q_prime, E_d_prime, E_fd, V_d, V_q, I_d, I_q, delta, omega, theta_vt = solution

# Display results
print(f"Steady-state results:")
print(f"Delta (rotor angle): {delta/np.pi*180:.4f} degree")
print(f"E'_q (q-axis transient internal voltage): {E_q_prime:.4f} pu")
print(f"E'_d (d-axis transient internal voltage): {E_d_prime:.4f} pu")
print(f"I_d (d-axis current): {I_d:.4f} pu")
print(f"I_q (q-axis current): {I_q:.4f} pu")
print(f"Omega (rotor speed): {omega:.4f} rad/s")
print(f"V_d (d-axis terminal voltage): {V_d:.4f} pu")
print(f"V_q (q-axis terminal voltage): {V_q:.4f} pu")
print(f"V_t (terminal voltage magnitude): {power_flow_results['V_t']:.4f} pu")
print(f"theta_vt (terminal voltage angle): {theta_vt/np.pi*180:.4f} degree")
```

Now assume a fault at bus 3 at time t=1.0, cleared by opening both lines into bus 3 at time = 1.1 seconds

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt


def two_axis_model_dynamic(t, y, machine_params, network_params, power_flow_results, E_fd, fault_scenario):
    r"""
    Two-Axis Model for synchronous machine connected with an infinity bus
    
    $$
    T'_{do} \frac{dE'_q}{dt} = -E'_q - (X_d - X'_d) I_d + E_{fd}
    $$

    $$
    T'_{qo} \frac{dE'_d}{dt} = -E'_d + (X_q - X'_q) I_q
    $$

    $$
    \frac{d\delta}{dt} = \omega - \omega_s
    $$

    $$
    \frac{2H}{\omega_s} \frac{d\omega}{dt} = P_M / (omega / omega_s) - E'_d I_d - E'_q I_q - (X'_q - X'_d) I_d I_q - T_{FW}
    $$

    $$
    0 = (R_s + R_e) I_d - (X'_q + X_{ep}) I_q - E'_d + V_s \sin(\delta - \theta_{vs})
    $$

    $$
    0 = (R_s + R_e) I_q + (X'_d + X_{ep}) I_d - E'_q + V_s \cos(\delta - \theta_{vs})
    $$

    $$
    V_d = R_{ed} I_d - X_{ep} I_q + V_s \sin(\delta - \theta_{vs})
    $$

    $$
    V_q = R_{eq} I_q + X_{ep} I_d + V_s \cos(\delta - \theta_{vs})
    $$

    $$
    V_d = V_t \sin(\delta - \theta_{vt})
    $$

    $$
    V_q = V_t \cos(\delta - \theta_{vt})
    $$
    """
      
    E_q_prime, E_d_prime, delta, omega = y
    
    H, X_d, X_q, X_d_prime, X_q_prime, T_do_prime, T_qo_prime, omega_s, T_FW = machine_params["H"], machine_params["X_d"], machine_params["X_q"], machine_params["X_d_prime"], machine_params["X_q_prime"], machine_params["T_do_prime"], machine_params["T_qo_prime"], machine_params["omega_s"], machine_params["T_FW"]
    P_M = power_flow_results["P_M"]
    
    # Determine network state based on fault scenario
    if t >= fault_scenario["fault_start"] and t < fault_scenario["fault_clear"]:
        network_params["X_ep"] = fault_scenario["X_ep_during_fault"]
    elif t >= fault_scenario["fault_clear"]:
        network_params["X_ep"] = fault_scenario["X_ep_after_clear"]
    
    def algebraic_equations(y, E_q_prime, E_d_prime, delta, machine_params, network_params, power_flow_results):
        I_d, I_q = y
        R_s, X_d_prime, X_q_prime = machine_params["R_s"], machine_params["X_d_prime"], machine_params["X_q_prime"]
        R_e, X_ep = network_params["R_e"], network_params["X_ep"]
        V_s, theta_vs = np.abs(power_flow_results["V_s"]), np.angle(power_flow_results["V_s"])
        
        eq1 = (R_s + R_e) * I_d - (X_q_prime + X_ep) * I_q - E_d_prime + V_s * np.sin(delta - theta_vs)
        eq2 = (R_s + R_e) * I_q + (X_d_prime + X_ep) * I_d - E_q_prime + V_s * np.cos(delta - theta_vs)
    
        return [eq1, eq2]

    # Solve for I_d and I_q
    I_d, I_q = fsolve(algebraic_equations, [0, 0], args=(E_q_prime, E_d_prime, delta, machine_params, network_params, power_flow_results))
    
    # Differential equations
    dE_q_prime_dt = (-E_q_prime - (X_d - X_d_prime) * I_d + E_fd) / T_do_prime
    dE_d_prime_dt = (-E_d_prime + (X_q - X_q_prime) * I_q) / T_qo_prime
    ddelta_dt = omega - omega_s
    domega_dt = (P_M / (omega / omega_s) - E_d_prime*I_d - E_q_prime*I_q - (X_q_prime - X_d_prime) * I_d * I_q - T_FW) * (omega_s / (2*H))
    
    return [dE_q_prime_dt, dE_d_prime_dt, ddelta_dt, domega_dt]

# Set up parameters
machine_params = {
    "H": 3.0,
    "R_s": 0,
    "X_d": 2.1,
    "X_q": 2.0,
    "X_d_prime": 0.3,
    "X_q_prime": 0.5,
    "T_do_prime": 7.0,
    "T_qo_prime": 0.75,
    "omega_s": 2*np.pi*60,
    "T_FW": 0,
}

network_params = {
    "R_e": 0,
    "X_ep": 0.20 * (0.10 + 0.20) / (0.20 + (0.10 + 0.20)) + 0.10,
}

power_flow_results = {
    "V_s": 1.0 + 1j*0,
    "V_t": 1.0946,
    "P_M": 1.0,
}

fault_scenario = {
    "fault_start": 1.0,
    "fault_clear": 1.1,
    "X_ep_during_fault": 0.10,  # Reduced impedance during fault (assume, not true equivalent impedance)
    "X_ep_after_clear": 0.30,   # Increased impedance after line opening
}

# Time span for the solution
t_span = (0, 20)

initial_guess = [0] * 10
initial_guess[-2] = 2*np.pi*60  # initialize the omega with non-zero to avoid dividing by zero warnings
# solve steady-state
solution = fsolve(two_axis_model_steady_state, initial_guess, args=(machine_params, network_params, power_flow_results))
E_q_prime, E_d_prime, E_fd, V_d, V_q, I_d, I_q, delta, omega, theta_vt = solution

# initial condition for machine state from steady-state
y0 = [E_q_prime, E_d_prime, delta, omega]

# Solve ODE
solution = solve_ivp(
    two_axis_model_dynamic,
    t_span,
    y0,
    args=(machine_params, network_params, power_flow_results, E_fd, fault_scenario),
    dense_output=True,
    method='RK45',
    max_step = 1e-3,
)

# Generate time points for plotting
t = np.linspace(0, 20, 1000)
y = solution.sol(t)

# Plot results
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(t, y[0])
plt.title("E'q")
plt.xlabel("Time (s)")
plt.ylabel("pu")

plt.subplot(2, 2, 2)
plt.plot(t, y[1])
plt.title("E'd")
plt.xlabel("Time (s)")
plt.ylabel("pu")

plt.subplot(2, 2, 3)
plt.plot(t, y[2] * 180 / np.pi)
plt.title("Delta")
plt.xlabel("Time (s)")
plt.ylabel("Degrees")

plt.subplot(2, 2, 4)
plt.plot(t, y[3] / (2*np.pi))
plt.title("Frequency")
plt.xlabel("Time (s)")
plt.ylabel("Hz")

plt.tight_layout()
plt.show()
```
